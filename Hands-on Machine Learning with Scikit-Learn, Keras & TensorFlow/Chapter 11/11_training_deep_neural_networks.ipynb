```python
from tensorflow import keras

model = keras.models.Sequential([
    # 1. Lapisan Input: Meratakan gambar 28x28 menjadi array 1D (784 fitur)
    keras.layers.Flatten(input_shape=[28, 28]),

    # 2. Lapisan Tersembunyi Pertama
    keras.layers.Dense(300, kernel_initializer="he_normal"),
    keras.layers.BatchNormalization(),
    keras.layers.Activation("relu"),

    # Loop untuk menambahkan 19 lapisan tersembunyi berikutnya
    # Setiap lapisan memiliki 100 neuron
    *[keras.layers.Dense(100, kernel_initializer="he_normal", use_bias=False) for _ in range(19)],
    *[keras.layers.BatchNormalization() for _ in range(19)],
    *[keras.layers.Activation("relu") for _ in range(19)],

    # 3. Lapisan Output: 10 neuron (untuk 10 kelas) dengan aktivasi softmax
    keras.layers.Dense(10, activation="softmax")
])

# Menampilkan ringkasan arsitektur model
model.summary()
```

```python
# Memuat model Xception yang sudah dilatih di ImageNet, tanpa lapisan atasnya
base_model = keras.applications.Xception(weights="imagenet", include_top=False)

# Membekukan bobot pada base model
for layer in base_model.layers:
    layer.trainable = False

# Membuat model baru di atas base model
model = keras.models.Sequential([
    base_model,
    keras.layers.GlobalAveragePooling2D(),
    keras.layers.Dense(10, activation="softmax") # 10 kelas untuk tugas baru
])
```

```python
# Menggunakan optimizer Adam saat kompilasi model
optimizer = keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss="sparse_categorical_crossentropy", optimizer=optimizer, metrics=["accuracy"])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi aktivasi
def leaky_relu(z, alpha=0.01):
    return np.maximum(alpha * z, z)

def elu(z, alpha=1):
    return np.where(z < 0, alpha * (np.exp(z) - 1), z)

def selu(z, scale=1.0507, alpha=1.6732):
    return scale * elu(z, alpha)

# Membuat data untuk plotting
z = np.linspace(-5, 5, 200)

# Plotting
plt.figure(figsize=(11, 4))

plt.subplot(121)
plt.plot(z, np.sign(z), "r-", linewidth=1, label="Step")
plt.plot(z, 1 / (1 + np.exp(-z)), "g--", linewidth=2, label="Sigmoid")
plt.plot(z, np.tanh(z), "b-", linewidth=2, label="Tanh")
plt.plot(z, np.maximum(0, z), "m-.", linewidth=2, label="ReLU")
plt.grid(True)
plt.legend(loc="center right", fontsize=14)
plt.title("Fungsi Aktivasi Umum")
plt.axis([-5, 5, -1.2, 1.2])

plt.subplot(122)
plt.plot(z, leaky_relu(z, 0.05), "b-", linewidth=2, label=r"Leaky ReLU ($\alpha=0.05$)")
plt.plot(z, elu(z), "g--", linewidth=2, label=r"ELU ($\alpha=1$)")
plt.plot(z, selu(z), "r-.", linewidth=2, label=r"SELU")
plt.grid(True)
plt.legend(loc="center right", fontsize=14)
plt.title("Varian ReLU")
plt.axis([-5, 5, -2.2, 2])

plt.show()
```

```python
# Fungsi scheduler yang berbeda
def power_scheduling(lr0, n_steps):
    return lambda epoch: lr0 / (1 + epoch / n_steps)**0.5

def exponential_scheduling(lr0, n_steps):
    return lambda epoch: lr0 * 0.1**(epoch / n_steps)

def piecewise_constant_scheduling(boundaries, values):
    boundaries = np.array([0] + boundaries)
    values = np.array(values)
    return lambda epoch: values[np.argmax(boundaries > epoch) - 1]

# Inisialisasi
n_epochs = 25
lr0 = 0.01
n_steps = 20
power_scheduler = power_scheduling(lr0, n_steps)
exp_scheduler = exponential_scheduling(lr0, n_steps)
piecewise_scheduler = piecewise_constant_scheduling([5, 15], [0.01, 0.005, 0.001])

# Menghitung learning rate untuk setiap epoch
lrs_power = [power_scheduler(epoch) for epoch in range(n_epochs)]
lrs_exp = [exp_scheduler(epoch) for epoch in range(n_epochs)]
lrs_piecewise = [piecewise_scheduler(epoch) for epoch in range(n_epochs)]

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(range(n_epochs), lrs_power, "b-", label="Power scheduling")
plt.plot(range(n_epochs), lrs_exp, "r--", label="Exponential scheduling")
plt.plot(range(n_epochs), lrs_piecewise, "g-.", label="Piecewise constant scheduling")
plt.xlabel("Epochs")
plt.ylabel("Learning Rate")
plt.title("Contoh Penjadwalan Learning Rate")
plt.grid(True)
plt.legend()
plt.show()
```
